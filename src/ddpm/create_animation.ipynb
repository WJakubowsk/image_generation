```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load image files from directory
from PIL import Image
import os
import re

def extract_number(filename):
    match = re.search(r'\d+', filename)
    return int(match.group()) if match else 0

images = []
dir = 'logs/DDPM_LSUN_L/sample'
image_list = os.listdir(dir)

image_list.sort(key=extract_number)

for filename in image_list:
    img = Image.open(dir + "/" + filename)
    images.append(img)
```

```python
import imageio

# Convert Pillow images to numpy arrays
image_arrays = [imageio.core.util.Array(np.array(image)) for image in images][:50]

# Create and save the GIF
output_path = 'bedroom_diffusion_animation.gif'
imageio.mimsave(output_path, image_arrays, duration=0.5)

print(f"GIF saved to {output_path}")
```
